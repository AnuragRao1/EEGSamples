```python
import pandas as pd
import numpy as np
import json
import csv

# Ratings
rating_data = pd.read_csv('/content/drive/MyDrive/neurosity_data/UserSessionData.csv') 
rating_df = pd.DataFrame(rating_data)
```

```python
#Get ratings as 10 x number_of_people array

# Convert string of array ratings into an array on int ratings
for i in range(len(rating_df)):
  array_num = json.loads(rating_df['ratings'].loc[i])
  rating_df['ratings'].loc[i] = [int(numeric_str) for numeric_str in array_num] 

# Dropping rows with less than 10 ratings
row_to_drop = []
for i in range(len(rating_df)):
  if len(rating_df['ratings'].loc[i]) != 10:
    row_to_drop.append(i)
rating_df.drop(row_to_drop, inplace = True)
rating_df = rating_df.reset_index()

rate = np.array(rating_df['ratings']) #shapes data into user x songs

# Want to tranpose into songs x user
rez = [[rate[j][i] for j in range(len(rate))] for i in range(len(rate[0]))]
rate = np.array(rez)
rating_df
```

```python
clf = RandomForestRegressor()
x_train, x_test, y_train, y_test = train_test_split(dataflat, ratingsflat, test_size = 0.70, random_state = 5)
clf.fit(x_train, y_train)
```

```python
y_pred_test = clf.predict(x_test)
diffs = []
for i in range(50):
    diffs.append( y_pred_test[i] - y_test[i] )
print(np.mean([x**2 for x in diffs])) #could be worse..
```

```python
clf = AdaBoostRegressor()
clf.fit(x_train, y_train)
y_pred_test = clf.predict(x_test)
diffs = []
for i in range(50):
    diffs.append( y_pred_test[i] - y_test[i] )
print(np.mean([x**2 for x in diffs]))
```

```python
clf = GradientBoostingRegressor()
clf.fit(x_train, y_train)
y_pred_test = clf.predict(x_test)
diffs = []
for i in range(50):
    diffs.append( y_pred_test[i] - y_test[i] )
print(np.mean([x**2 for x in diffs]))
```

```python
from sklearn import linear_model
clf = linear_model.LogisticRegression()
clf.fit(x_train, y_train)
y_pred_test = clf.predict(x_test)
diffs = []
for i in range(50):
    diffs.append( y_pred_test[i] - y_test[i] )
print(np.mean([x**2 for x in diffs]))
```

```python
from sklearn import neighbors
clf = neighbors.KNeighborsRegressor()
clf.fit(x_train, y_train)
y_pred_test = clf.predict(x_test)
diffs = []
for i in range(50):
    diffs.append( y_pred_test[i] - y_test[i] )
print(np.mean([x**2 for x in diffs]))
```
